# Life Expectancy vs. GDP

This Python notebook will analyze the data provided by Codeacademy for the Life Expectancy vs. GDP project. The data file contains the following information

- Year (between 2000-2015)
- Country (6 countries in total)
- Life Expectancy for a given country for a given year
- GDP for each of the 6 countries each year

The goal of this project will be to examine how the Life Expectancy and GDP of each country evolves with time and determine if there is any correlation between the two. I would guess that as GDP increases, so too would the Life Expectancy.

Before beginning, it is important to note that the dataset provided does not provide gender specific information for each country (i.e. life expectancy based on gender identity). As such, it will be impossible to draw gender specific conclusions based on these data, which could be important.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
LE_v_GDP_df = pd.read_csv('all_data.csv') ### creating the life expectancy and GDP data frame object
print(LE_v_GDP_df.head()) ### printing the first 5 rows of the column to get a sense of the data set
```

```python
'''The GDP column looks messy with it being in scientfic notation. 
   I will add a column to display this data in billions.'''

LE_v_GDP_df['GDP_in_billions'] = LE_v_GDP_df['GDP'] / 10 ** 9
print(LE_v_GDP_df.head()) ### again, just printing the first 5 rows to get a sense of the new df.
```

```python
'''
It looks like a line graph will do a good job of illustrating the 
time evolution of life expectancy and GDP for each country.
'''

def time_evo_plot(column_name,data_frame):
    '''
    This function will plot the time evolution for a given
    column in the data frame.
    
    column_name = the name of the column you wish to 
                  investigate.
                  
    data_frame = the data frame containing the data.
    '''
    countries = []
    for country in data_frame['Country']. unique():
        country_df = data_frame[data_frame['Country'] == country]
        years = country_df['Year'].values
        column_data = country_df[column_name].values
        plt.plot(years, column_data, marker = '^')
        countries.append(country)
    plt.xlabel('Year')
    if column_name == 'Life expectancy at birth (years)':
        plt.ylabel('Life Expectancy (years)')
        plt.title('Time Evolution of Life Expectancy at Birth for 6 Countries')
    elif column_name == 'GDP_in_billions':
        plt.ylabel('GDP in Billions ($USD)')
        plt.title('Time Evolution of GDP in Billions USD for 6 Countries')
    plt.legend(countries)

ax = plt.subplots(figsize = (15,8))
plt.subplot(1,2,1)
life_expectancy = time_evo_plot('Life expectancy at birth (years)', LE_v_GDP_df)

plt.subplot(1,2,2)
gdp_evo = time_evo_plot('GDP_in_billions', LE_v_GDP_df)

plt.savefig(r'C:\Users\zcron\Desktop\LE_v_GDP_Project\GDP_LE_year.png', format = 'png') 

plt.show()
```

```python
def gdp_v_le(country, data_frame):
    '''
    This function plots the LE as a function of GDP for 
    each country.
    
    data_frame = The data frame containing the data.
    
    country = the name of the country.
    '''
    country_df = data_frame[data_frame['Country'] == country]
    le_data = country_df['Life expectancy at birth (years)'].values
    gdp_data = country_df['GDP_in_billions'].values
    plt.scatter(gdp_data, le_data, marker = '^')
    plt.xlabel('GDP in Billions ($USD)')
    plt.ylabel('Life Expectancy at Birth (years)')
    if country != 'United States of America':
        plt.title('Life Expectancy vs. GDP of ' + country + ', 2000-2015')
    else:
        plt.title('Life Expectancy vs. GDP of U.S.A., 2000-2015')
        
ax = plt.subplots(figsize=(15,15))
for i in range(len(LE_v_GDP_df['Country'].unique())):
    plt.subplot(3, 2 , i + 1)
    gdp_v_le(LE_v_GDP_df['Country'].unique()[i], LE_v_GDP_df)

plt.subplots_adjust(wspace = 0.25, hspace = 0.3)

plt.savefig(r'C:\Users\zcron\Desktop\LE_v_GDP_Project\GDP_v_LE.png', format = 'png') 

plt.show()
```

```python
def LE_hist(data_frame):
    '''
    This function will plot the histogram of life 
    expectancies for the total data set. It will 
    then over plot a vertical line where the mean 
    and median life expectancies are located.
    
    data_frame = the data_frame containing the 
                 relavent data.
    '''
    le_data = data_frame['Life expectancy at birth (years)'].values
    le_mean = np.mean(le_data)
    le_median = np.median(le_data)
    plt.hist(le_data, range = (int(np.min(le_data)) - 1, int(np.max(le_data)) + 1), bins = 50, edgecolor = 'black')
    plt.axvline(le_mean, color = 'red', linestyle = '--', label = 'Mean: {mean: .2f}'.format(mean = le_mean))
    plt.axvline(le_median, color = 'orange', linestyle = '--', label = 'Median: {median: .2f}'.format(median = le_median))
    plt.xlabel('Life Expectancy at Birth (years)')
    plt.ylabel('Counts')
    plt.title('Life Expectancy Distribution for 6 Countries Between 2000-2015')

plt.figure(figsize = (10,8))
LE_hist(LE_v_GDP_df)
plt.legend()

plt.savefig(r'C:\Users\zcron\Desktop\LE_v_GDP_Project\LE_hist.png', format = 'png') 

plt.show()
```

```python
def le_violin(country, data_frame):
    '''
    This function will create a violin plot of 
    the life expectancies for each of the 6 
    countries.
    
    data_frame = the data frame containing the
                 relevant data.
    '''
    country_df = data_frame[data_frame['Country'] == country]
    sns.violinplot(
        x = 'Country',
        y = 'Life expectancy at birth (years)',
        data = country_df,
        pallete = 'Set1')
    if country != 'United States of America':
        plt.title('Life Expectancy Distributions for ' + country + ' Between 2000-2015')
    else:
        plt.title('Life Expectancy Distributions for U.S.A. Between 2000-2015')
    
plt.subplots(figsize = (15,15))   
for i in range(len(LE_v_GDP_df['Country'].unique())):
    ax = plt.subplot(3,2, i + 1)
    le_violin(LE_v_GDP_df['Country'].unique()[i], LE_v_GDP_df)
    ax.set_xticklabels(' ')
    plt.xlabel(' ')
plt.subplots_adjust(hspace = 0.25)

plt.savefig(r'C:\Users\zcron\Desktop\LE_v_GDP_Project\LE_violin.png', format = 'png') 

plt.show()
```
